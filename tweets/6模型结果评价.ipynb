```python
import pandas as pd
tweet=pd.read_csv("D:/安装包/LDA/LDA.txt",sep='\t',names=['date','text','loc','author','class'])
tweet.head()
```

```python
import plotly.graph_objects as go 

#绘制真实虚假灾难数量对比的饼图
counts_train = tweet['class'].value_counts(sort=False) #查看target列中有多少个不同值并计算每个不同值有在该列中有多少重复值
labels = counts_train.index   
values_train = counts_train.values 

data = go.Pie(labels=["Topic0","Topic1","Topic2","Topic3"], values=values_train ,pull=[0.03, 0])                  #输入饼图数据
layout = go.Layout(title='Determining which type of tweet belongs (%) ') #设置标题

fig = go.Figure(data=[data], layout=layout)                                        #绘制饼图
fig.update_traces(hole=.3, hoverinfo="label+percent+value")                        #设置饼图的显示，标签，百分比，值
fig.update_layout(annotations=                                                     #在饼图的中心添加注释
                  [dict(text='Train', x=0.5, y=0.5, font_size=20, showarrow=False)])
fig.show()                                                                         #显示饼图
```

```python
import seaborn as sns
import matplotlib.pyplot as plt  

tweet['date'] = pd.to_datetime(tweet['date'])
tweet['Month'] = tweet['date'].dt.month
tweet['class'] = tweet['class'].astype('str')

counts_0 = tweet[tweet['class']=='0']['Month'].value_counts(sort=False)
labels = counts_0.index  
values_0= counts_0.values 
counts_1 = tweet[tweet['class']=='1']['Month'].value_counts(sort=False)
values_1 = counts_1.values
counts_2 = tweet[tweet['class']=='2']['Month'].value_counts(sort=False)
values_2 = counts_2.values
counts_3 = tweet[tweet['class']=='3']['Month'].value_counts(sort=False)
values_3 = counts_3.values
print(labels)
data={"Month":labels,"Topic0":values_0,"Topic1":values_1,"Topic2":values_2,"Topic3":values_3}
df = pd.DataFrame(data, labels, ["Topic0","Topic1","Topic2","Topic3"])
ax=sns.lineplot(data=df)
ax.set_title("Trends in Tweet Topics")
plt.show()
```
